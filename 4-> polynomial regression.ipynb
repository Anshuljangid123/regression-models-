# polynomial linear regression .

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/content/Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

# training the linear regression model  -> on the whole dataset.

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
x_reshaped = x.reshape(-1,1)
lin_reg.fit(x_reshaped,y) # here to train the model we are giving the whole dataset.
```

# Implementing the polynomial regression model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
# Fit: It analyzes your existing feature x to understand its structure.
#Transform: It creates new polynomial features based on the degree you specified.
# For example, if x was [1, 2, 3], x_poly might become [1, 1, 2, 4, 3, 9] (including the constant term, x, and x^2).
x_poly = poly_reg.fit_transform(x)

# making the object for the polynimial regression .

#Creating a Linear Regression Object for Polynomial Regression:
lin_reg_2 = LinearRegression()
# trains your linear regression model on the newly created polynomial features x_poly and the target values y.
# This finds the best coefficients for the polynomial equation that predicts y based on the polynomial features of x.

lin_reg_2.fit(x_poly , y)
```

# visualizing the linear regression results

```python
plt.scatter(x,y , color = 'red')

#This part uses the trained linear regression model (lin_reg) to predict the corresponding salary values for each position level in x.
plt.plot(x, lin_reg.predict(x) , color = 'Blue')
plt.title('truth or bluff (linear regression .)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x,y , color = 'red')
plt.plot(x, lin_reg_2.predict(x_poly) , color = 'Blue')
plt.title('truth or bluff (polynomial linear regression .)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

# We will retrain the polynomial regression model with the higher degree.

```python
poly_reg = PolynomialFeatures(degree = 4)
# Fit: It analyzes your existing feature x to understand its structure.
#Transform: It creates new polynomial features based on the degree you specified.
# For example, if x was [1, 2, 3], x_poly might become [1, 1, 2, 4, 3, 9] (including the constant term, x, and x^2).
x_poly = poly_reg.fit_transform(x)

# making the object for the polynimial regression .

#Creating a Linear Regression Object for Polynomial Regression:
lin_reg_2 = LinearRegression()
# trains your linear regression model on the newly created polynomial features x_poly and the target values y.
# This finds the best coefficients for the polynomial equation that predicts y based on the polynomial features of x.

lin_reg_2.fit(x_poly , y)
```

```python
plt.scatter(x,y , color = 'red')
plt.plot(x, lin_reg_2.predict(x_poly) , color = 'Blue')
plt.title('truth or bluff (polynomial linear regression -> with degree 4.)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

### to get the smoother curve we need to increase the density of the x axis numbers

```python
x_grid = np.arange(min(x), max(x), 0.1)
'''#This line creates an array X_grid that contains evenly spaced values from
 the minimum to the maximum of your feature data X (position levels).
 The spacing between these values is 0.1. This denser grid helps create a smoother curve for
  the polynomial regression line.'''

x_grid = x_grid.reshape((len(x_grid), 1))
#This line reshapes the X_grid array into a column vector.
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
